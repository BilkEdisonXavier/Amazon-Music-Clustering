```python
#import libraries for clustering
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import dendrogram, linkage

# 1. Standardize the dataset
numeric_cols = data.select_dtypes(include=[np.number]).columns
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data[numeric_cols])

# 2. Dimensionality Reduction (PCA 2D for visualization)

pca = PCA(n_components=2)
pca_result = pca.fit_transform(data_scaled)

# 3A. KMeans Clustering

print("=== KMeans Clustering ===")
sse = []
for k in range(2, 11):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(data_scaled)
    sse.append(km.inertia_)

# Elbow method plot
plt.figure(figsize=(8,5))
plt.plot(range(2,11), sse, 'bo-')
plt.title("KMeans Elbow Method")
plt.xlabel("k")
plt.ylabel("SSE (Inertia)")
plt.show()

# Example: Choose k=4
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans_labels = kmeans.fit_predict(data_scaled)
data["KMeans_Cluster"] = kmeans_labels

print("Silhouette Score (KMeans, k=4):", silhouette_score(data_scaled, kmeans_labels))

plt.figure(figsize=(8,6))
plt.scatter(pca_result[:,0], pca_result[:,1], c=kmeans_labels, cmap='viridis', alpha=0.6)
plt.title("KMeans Clustering (PCA 2D)")
plt.xlabel("PC1"); plt.ylabel("PC2")
plt.colorbar(label="Cluster")
plt.show()

# 3B. DBSCAN Clustering

print("=== DBSCAN Clustering ===")
dbscan = DBSCAN(eps=1.5, min_samples=5)  # tune eps & min_samples
dbscan_labels = dbscan.fit_predict(data_scaled)
data["DBSCAN_Cluster"] = dbscan_labels

n_clusters = len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0)
print(f"DBSCAN found {n_clusters} clusters (+ noise).")

plt.figure(figsize=(8,6))
plt.scatter(pca_result[:,0], pca_result[:,1], c=dbscan_labels, cmap='plasma', alpha=0.6)
plt.title("DBSCAN Clustering (PCA 2D)")
plt.xlabel("PC1"); plt.ylabel("PC2")
plt.colorbar(label="Cluster (-1 = Noise)")
plt.show()
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score

# Evaluation Function

def evaluate_clustering(X, labels, method="KMeans"):
    # Skip if clustering created only 1 cluster (e.g., DBSCAN sometimes)
    if len(set(labels)) <= 1:
        print(f"{method}: Only one cluster found. Evaluation skipped.")
        return
    
    sil = silhouette_score(X, labels)
    dbi = davies_bouldin_score(X, labels)
    
    print(f"=== {method} Evaluation ===")
    print(f"Silhouette Score: {sil:.4f} (closer to 1 is better)")
    print(f"Davies-Bouldin Index: {dbi:.4f} (lower is better)")
    if method == "KMeans":
        print(f"Inertia (SSE): {kmeans.inertia_:.4f} (lower = compact clusters)")
    print()

# Evaluate Each Method

evaluate_clustering(data_scaled, kmeans_labels, "KMeans")
evaluate_clustering(data_scaled, dbscan_labels, "DBSCAN")
```

```python
# Cluster Profiling Function

def profile_clusters(df, cluster_col, numeric_cols):
    print(f"\n=== Cluster Profiling: {cluster_col} ===")
    
    # Group by cluster and calculate mean feature values
    profile = df.groupby(cluster_col)[numeric_cols].mean().round(2)
    
    # Show profile table
    display(profile)
    
    # Quick heatmap for visualization
    plt.figure(figsize=(10,6))
    sns.heatmap(profile, annot=True, cmap="coolwarm", cbar=True)
    plt.title(f"Cluster Profiling Heatmap ({cluster_col})")
    plt.show()

# Apply Profiling for Each Method
profile_clusters(data, "KMeans_Cluster", numeric_cols)
profile_clusters(data, "DBSCAN_Cluster", numeric_cols)
```

```python
# === Unsupervised Clustering with Train/Test Split ===

import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score
from sklearn.datasets import make_blobs

# 1. Example dataset

X, _ = make_blobs(n_samples=1000, n_features=2, centers=3, cluster_std=1.0, random_state=42)

# Train/Test split (even though unsupervised has no labels, we test stability)
X_train, X_test = train_test_split(X, test_size=0.2, random_state=42)

# 2. KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_train)

# Predict on test set
kmeans_labels = kmeans.predict(X_test)

# Evaluate on test set
kmeans_sil = silhouette_score(X_test, kmeans_labels)
kmeans_dbi = davies_bouldin_score(X_test, kmeans_labels)

print("=== KMeans Test Evaluation ===")
print(f"Silhouette Score: {kmeans_sil:.4f} (closer to 1 is better)")
print(f"Davies-Bouldin Index: {kmeans_dbi:.4f} (lower is better)")
print("-"*50)

# 3. DBSCAN

dbscan = DBSCAN(eps=0.8, min_samples=5)   # tune eps & min_samples
dbscan.fit(X_train)

# DBSCAN doesn't support predict, so we fit directly on test set too
dbscan_test = DBSCAN(eps=0.8, min_samples=5)
dbscan_labels = dbscan_test.fit_predict(X_test)

# Evaluate only if more than 1 cluster found
print("=== DBSCAN Test Evaluation ===")
if len(set(dbscan_labels)) > 1:
    dbscan_sil = silhouette_score(X_test, dbscan_labels)
    dbscan_dbi = davies_bouldin_score(X_test, dbscan_labels)
    print(f"Silhouette Score: {dbscan_sil:.4f} (closer to 1 is better)")
    print(f"Davies-Bouldin Index: {dbscan_dbi:.4f} (lower is better)")
else:
    print("DBSCAN found only one cluster (or mostly noise).")
```

```python
# === Unsupervised Clustering with Train/Test Split + Visualization ===

import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score
from sklearn.datasets import make_blobs

# 1. Example dataset

X, _ = make_blobs(n_samples=1000, n_features=2, centers=3, cluster_std=1.0, random_state=42)

# Train/Test split
X_train, X_test = train_test_split(X, test_size=0.2, random_state=42)


# 2. KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_train)

# Predict on train & test
kmeans_labels_train = kmeans.labels_
kmeans_labels_test = kmeans.predict(X_test)

# Evaluate on test set
kmeans_sil = silhouette_score(X_test, kmeans_labels_test)
kmeans_dbi = davies_bouldin_score(X_test, kmeans_labels_test)

print("=== KMeans Test Evaluation ===")
print(f"Silhouette Score: {kmeans_sil:.4f} (closer to 1 is better)")
print(f"Davies-Bouldin Index: {kmeans_dbi:.4f} (lower is better)")
print("-"*50)

# 3. DBSCAN

dbscan = DBSCAN(eps=0.8, min_samples=5)
dbscan_labels_train = dbscan.fit_predict(X_train)

# Fit on test set separately
dbscan_test = DBSCAN(eps=0.8, min_samples=5)
dbscan_labels_test = dbscan_test.fit_predict(X_test)

print("=== DBSCAN Test Evaluation ===")
if len(set(dbscan_labels_test)) > 1:
    dbscan_sil = silhouette_score(X_test, dbscan_labels_test)
    dbscan_dbi = davies_bouldin_score(X_test, dbscan_labels_test)
    print(f"Silhouette Score: {dbscan_sil:.4f} (closer to 1 is better)")
    print(f"Davies-Bouldin Index: {dbscan_dbi:.4f} (lower is better)")
else:
    print("DBSCAN found only one cluster (or mostly noise).")
print("-"*50)

# 4. Visualization

fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# KMeans Train
axs[0, 0].scatter(X_train[:, 0], X_train[:, 1], c=kmeans_labels_train, cmap="viridis", s=20)
axs[0, 0].set_title("KMeans - Train Clusters")

# KMeans Test
axs[0, 1].scatter(X_test[:, 0], X_test[:, 1], c=kmeans_labels_test, cmap="viridis", s=20)
axs[0, 1].set_title("KMeans - Test Clusters")

# DBSCAN Train
axs[1, 0].scatter(X_train[:, 0], X_train[:, 1], c=dbscan_labels_train, cmap="plasma", s=20)
axs[1, 0].set_title("DBSCAN - Train Clusters")

# DBSCAN Test
axs[1, 1].scatter(X_test[:, 0], X_test[:, 1], c=dbscan_labels_test, cmap="plasma", s=20)
axs[1, 1].set_title("DBSCAN - Test Clusters")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans
import pickle

# Load your dataset
df = pd.read_csv("D:\Projects\Project-4-Amazon Music Clustering\data_preprocessed.csv")

# Drop ID/URL columns
df = df.drop(columns=[col for col in df.columns if "id" in col.lower() or "url" in col.lower()])

# Encode categorical features
for col in df.select_dtypes(include=["object"]).columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))

# Features used for training
training_columns = df.columns.tolist()

# Train KMeans
model = KMeans(n_clusters=3, random_state=42)
model.fit(df)

# ---- Save Model ----
with open("kmeans_model.pkl", "wb") as f:
    pickle.dump(model, f)

print("KMeans model saved successfully!")
# Save model + training columns together
pickle.dump((model, training_columns), open("model_with_columns.pkl", "wb"))
print("Model with training columns saved!")
```
